```python
from huggingface_hub import hf_hub_download
import os

MODEL_PATH = '/content/sd_xl_base_1.0.safetensors'

if not os.path.exists(MODEL_PATH):
    print("İndiriliyor (~7 GB), bekle...")
    hf_hub_download(
        repo_id="stabilityai/stable-diffusion-xl-base-1.0",
        filename="sd_xl_base_1.0.safetensors",
        local_dir="/content"
    )
    print("✅ İndirildi!")
else:
    print("✅ Model zaten var, atlandı.")
```

```python
config = f"""
[general]
enable_bucket = true

[[datasets]]
resolution = 1024
batch_size = 1

  [[datasets.subsets]]
  image_dir = "{TRAIN_DIR}/10_face"
  caption_extension = ".txt"
  num_repeats = 1
"""

config_path = '/content/dataset_config.toml'
with open(config_path, 'w') as f:
    f.write(config)

print("✅ Config yazıldı!")
print(config)
```

```python
import subprocess

cmd = [
    "accelerate", "launch",
    "--num_cpu_threads_per_process=4",
    "/content/sd-scripts/sdxl_train_network.py",
    f"--pretrained_model_name_or_path={MODEL_PATH}",
    f"--dataset_config={config_path}",
    f"--output_dir={OUTPUT_DIR}",
    "--output_name=celeba_hq_lora_xl",
    "--network_module=networks.lora",
    "--network_dim=32",
    "--network_alpha=16",
    "--optimizer_type=AdamW8bit",
    "--learning_rate=1e-4",
    "--lr_scheduler=cosine_with_restarts",
    "--lr_warmup_steps=500",
    "--max_train_steps=10000",
    "--save_every_n_steps=2000",
    "--save_model_as=safetensors",
    "--mixed_precision=fp16",
    "--gradient_checkpointing",
    "--train_batch_size=1",
    "--cache_latents",
    "--cache_latents_to_disk",
    "--xformers",
    "--noise_offset=0.05",
]

print("🚀 SDXL LoRA eğitimi başlıyor!")
print(f"   Model: sd_xl_base_1.0")
print(f"   Steps: 10,000")
print(f"   Çözünürlük: 1024x1024")
print(f"   Output: {OUTPUT_DIR}\n")

process = subprocess.Popen(
    cmd,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True
)
for line in process.stdout:
    print(line, end='')
process.wait()
print("\n✅ Eğitim tamamlandı!")
```

```python
import glob
import os
checkpoints = sorted(glob.glob(os.path.join(OUTPUT_DIR, '*.safetensors')))
print(f"📦 {len(checkpoints)} checkpoint bulundu:\n")
for i, ckpt in enumerate(checkpoints):
    size_mb = os.path.getsize(ckpt) / 1024 / 1024
    print(f"  [{i}] {os.path.basename(ckpt)}  ({size_mb:.0f} MB)")
```

```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

LORA_PATH = checkpoints[-1]
print(f"Kullanılan: {os.path.basename(LORA_PATH)}")

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    safety_checker=None
).to("cuda")

pipe.load_lora_weights(LORA_PATH)
print("✅ Model yüklendi! Üretiliyor...")

prompts = [
    "photo of a young woman, blonde hair, smiling, high cheekbones",
    "photo of a middle-aged man, black hair, goatee, wearing glasses",
    "photo of a young woman, brown hair, wavy hair, with bangs, wearing lipstick",
    "photo of a young man, bald, smiling, mouth slightly open",
]

fig, axes = plt.subplots(1, len(prompts), figsize=(5 * len(prompts), 5))
for ax, prompt in zip(axes, prompts):
    image = pipe(
        prompt=prompt,
        negative_prompt="blurry, bad quality, cartoon, anime, deformed, ugly",
        num_inference_steps=30,
        guidance_scale=7.5,
        width=512, height=512,
    ).images[0]
    ax.imshow(image)
    ax.set_title(prompt[:40] + "...", fontsize=8)
    ax.axis('off')

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'generated_faces.png'), dpi=150)
plt.show()
print("✅ Görseller Drive'a kaydedildi!")
```
